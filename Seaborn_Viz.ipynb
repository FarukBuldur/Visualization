```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
fifa_filepath = "fifa.csv"

# Read the file into a variable fifa_data
fifa_data = pd.read_csv(fifa_filepath, index_col="Date", parse_dates=True)
```

```python
fifa_data.head()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(16,6))

# Line chart showing how FIFA rankings evolved over time 
sns.lineplot(data=fifa_data)
plt.show()
```

```python
# Path of the file to read
flight_filepath = "flight_delays.csv"

# Read the file into a variable flight_data
flight_data = pd.read_csv(flight_filepath, index_col="Month")
```

```python
flight_data.head()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=flight_data.index, y=flight_data['NK'])

# Add label for vertical axis
plt.ylabel("Arrival delay (in minutes)")
plt.show()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add title
plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=flight_data, annot=True)

# Add label for horizontal axis
plt.xlabel("Airline")
plt.show()
```

```python
# Path of the file to read
insurance_filepath = "insurance.csv"

# Read the file into a variable insurance_data
insurance_data = pd.read_csv(insurance_filepath)
```

```python
insurance_data.head()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(16,6))
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
plt.show()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(16,6))
# Add a Regression line to see the correlation btw bmi and charges
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
plt.show()
```

```python
#plt.figure(figsize=(16,6))
#Smoker & Non-Smoker People, we see that smoking is highly correlation with insurance charges
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
plt.show()
```

```python
#plt.figure(figsize=(16,6))
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
plt.show()
```

```python
#Swarm Plot 
plt.figure(figsize=(16,6))
sns.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['charges'])
plt.show()
```
